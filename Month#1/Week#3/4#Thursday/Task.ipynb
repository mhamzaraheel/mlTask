```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Load the data set
data = pd.read_csv("Trading-features.csv")
```

```python
data
```

```python
# Give all the data information
data.describe(include = 'all')
```

```python
# check the length and shape of the data
print(f"There are {len(data)} rows in the dataset.")
print(f"Shape of Data is {data.shape}",) # >>> (rows , columns)
```

```python
# Convert the date column into date time
data['created_at'] = pd.to_datetime(data['created_at'])
# Extracting the time part
data['created_at'] = data['created_at'].dt.time
```

```python
print('Columns with null values and their counts:')
null_columns = data.columns[data.isnull().any()]
for column in null_columns:
    null_count = data[column].isnull().sum()
    null_percentage = (null_count / len(data)) * 100
    print(f'{column}: {null_count} null values ({null_percentage:.2f}%)')
```

```python
# Remove columns containing null values and ensure that the removal does not significantly reduce the amount of data.
data.dropna(inplace=True)
len(data)
```

```python
# Columns with all values zero , remove all that column
zero_columns = [column for column in data.columns if (data[column] == 0).all()]
print("Columns with all zero values:", zero_columns)
data = data.drop(zero_columns, axis=1)
data.shape
```

```python
# Drop the row with all zeros values
data = data.loc[(data != 0).any(axis=1)]
data.shape
```

```python
# Drop the column <<id>>
data.drop("id", axis=1, inplace=True)
data.shape
```

```python
data
```

```python
# Our data set  <<Features>> <<Labels>>
x = data.iloc[:, 1:data.shape[1]]
y = data.iloc[:, 0]
x.shape , y.shape
```

```python
chosen_feature_index = 1
```

```python
# Assuming chosen_feature_index is the index of the feature you want to plot
chosen_feature = x.iloc[:, chosen_feature_index]

# Choose a colormap (e.g., 'viridis', 'plasma', 'magma', 'inferno', etc.)
colormap = 'viridis'

# Scatter plot with colormap
plt.scatter(chosen_feature, y, c=y, cmap=colormap)
plt.xlabel(x.columns[chosen_feature_index])
plt.ylabel('y')
plt.title('Data Before the Normalization')
plt.colorbar(label='Color scale')
plt.show()
```
